```python
https://github.com/TDO196/Agent_ThucHanh_TriTueNhanTao/tree/main
from collections import deque
def print_board(state):
    for i in range(0, 9, 3):
        print(state[i:i+3])
    print()
def get_neighbors(state):
    neighbors = []

    # vị trí của ô trống
    zero_index = state.index(0)

    # các bước di chuyển
    moves = {
        "UP": -3,
        "DOWN": 3,
        "LEFT": -1,
        "RIGHT": 1
    }

    row = zero_index // 3
    col = zero_index % 3

    for move, step in moves.items():
        # rule hợp lệ
        if move == "UP" and row == 0:
            continue
        if move == "DOWN" and row == 2:
            continue
        if move == "LEFT" and col == 0:
            continue
        if move == "RIGHT" and col == 2:
            continue
        # tạo trạng thái mới
        new_index = zero_index + step
        new_state = list(state)
        # hoán đổi
        new_state[zero_index], new_state[new_index] = (
            new_state[new_index],
            new_state[zero_index]
        )
        neighbors.append((move, tuple(new_state)))
    return neighbors
# Thuật toán BFS
def bfs(start, goal):

    frontier = deque()
    frontier.append((start, []))

    reached = set()
    reached.add(start)

    while frontier:

        current_state, path = frontier.popleft()

        # in trạng thái hiện tại
        print("Node đang xét:")
        print_board(current_state)

        # kiểm tra goal
        if current_state == goal:
            return path

        # sinh node con
        for move, neighbor in get_neighbors(current_state):

            if neighbor not in reached:
                reached.add(neighbor)

                new_path = path + [move]

                frontier.append((neighbor, new_path))

    return None
start_state = (1, 2, 3,
               4, 5, 6,
               7, 0, 8)

goal_state = (1, 2, 3,
              4, 5, 6,
              7, 8, 0)

print("Trạng thái ban đầu:")
print_board(start_state)
solution = bfs(start_state, goal_state)
if solution:
    print("Đã tìm thấy lời giải!")
    print("Số bước:", len(solution))
    print("Đường đi:", solution)
else:
    print("Không tìm thấy lời giải.")
```
